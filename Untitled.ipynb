```python
JSON_PATH = '/root/projects/supertrend'
super_trend_runner = SuperTrendRunner(ticker='VTI', json_path=JSON_PATH, multiplier=2.5, debug_mode=True, rolling_period=100, initial_amount=10000)
super_trend_df = super_trend_runner.generate_super_trend_for_ticker(period="5y")
```

```python
portfolio_over_time, percent_differences = super_trend_runner.simulate_portfolio_on_strategy(super_trend_df)

super_trend_df['portfolio_values'] = portfolio_over_time
super_trend_df['percentage_change'] = percent_differences

benchmark_comparison_df, benchmark_df = super_trend_runner.benchmark_strategy(super_trend_df)
super_trend_runner.visualize_benchmark_against_strategy(super_trend_df, benchmark_df)
```

```python
from collections import deque

buying_periods = deque()
current_day = 1
previous_day = 0
buying = False

while current_day != len(super_trend_df.index):
    previous_day_row = super_trend_df.iloc[previous_day]
    current_day_row = super_trend_df.iloc[current_day]
    
    previous_price = previous_day_row['Close']
    current_price = current_day_row['Close']
    
    # buying
    if current_day_row['buy_or_sell'] == True and previous_day_row['buy_or_sell'] == False:
        buying_periods.append([current_day, -1])
        buying = True
    
    # selling 
    elif current_day_row['buy_or_sell'] == False and previous_day_row['buy_or_sell'] == True:
        buying_periods[-1][1] = current_day
        buying = False
        
    previous_day += 1
    current_day += 1
```

```python
for bp in buying_periods:
    print(bp)
    bp[0] += 1
    bp[1] += 1
```

```python
i = 1
bp = buying_periods.popleft() if buying_periods else None
percentage_differences = [0.0]

while i != len(super_trend_df.index):
    raw_percentage = 0
    if bp:
        if i < bp[0]:
            raw_percentage = 0
        elif i >= bp[0] and i <= bp[1]:
            previous_price = super_trend_df.iloc[i - 1]['Close']
            current_price = super_trend_df.iloc[i]['Close']
            raw_percentage = ((current_price - previous_price) / previous_price)
        elif i > bp[0]:
            bp = buying_periods.popleft() if buying_periods else None
            
    percent_difference = 1 + raw_percentage
    percentage_differences.append(percent_difference)
    i += 1
```
