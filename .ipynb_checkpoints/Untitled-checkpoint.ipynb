```python
from super_trend_lib import SuperTrendRunner
```

```python
runner = SuperTrendRunner('AAPL', debug_mode=True)
super_trend_df = runner.run()
```

```python
super_trend_df.tail(10)
```

```python
i = len(super_trend_df.index) - 1
recent = None
while i > 0:
    current = super_trend_df.iloc[i]
    previous = super_trend_df.iloc[i - 1]
    
    
    
    percent_change = (100 + current['percentage_change']) / 100
    calculated = current['portfolio_values'] / percent_change
    original = previous['portfolio_values']
    
    if calculated == original:
        print("valid")
    else:
        print("invalid")
        print(f'calc: {calculated}')
        print(f'orig: {original}')
    i -= 1
```
